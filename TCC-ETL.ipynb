# Gerar a Correlação

### [Checkpoint] - Rodar as Bibliotecas antes de rodar as células de baixo

```python
dataset = pd.read_csv('/work/datasetCorrigido.csv')
```

```python
def covarianca(x, y):
    n = len(x)
    x_m = x - np.mean(x)
    y_m = y - np.mean(y)
    return (x_m * y_m).sum() / (n - 1)

def correlacao(x, y):
    cov_xy = covarianca(x, y)
    dp_x = np.std(x, ddof=1)
    dp_y = np.std(y, ddof=1)
    return cov_xy/(dp_x*dp_y)
```

```python
df_1 = _deepnote_execute_sql('select distinct(CODIGO) from dataset;', 'SQL_DEEPNOTE_DATAFRAME_SQL', audit_sql_comment='', sql_cache_mode='cache_disabled')
df_1
```

```python
corrColunas = [x for x in df_1['CODIGO']]
correlacaoDF = pd.DataFrame(columns=corrColunas)

i_correlacao = pd.DataFrame({"index": corrColunas})

correlacaoDF = correlacaoDF.append(i_correlacao)

correlacaoDF = correlacaoDF.set_index('index')

correlacaoDF = correlacaoDF.fillna(0)
```

```python
w = 0

for x in corrColunas:
    z = 0
    w = w + 1;
    #Cria a query do primeiro DataFrame.
    aux = 'CODIGO == "'+ str(x) +'"'
    df = dataset.query(aux).reset_index(drop=True)

    for y in corrColunas:
        if(w <= z):
            #Cria a query do segundo DataFrame.
            aux2 = 'CODIGO == "'+ str(y) +'"'
            df2 = dataset.query(aux2).reset_index(drop=True)
            #Calcula a correlacao.
            value_correlacao = correlacao(df['VALUE'], df2['VALUE'])
        
            #Adiciona na tabela.
            correlacaoDF.loc[x, y] = value_correlacao
        else:
            correlacaoDF.loc[x, y] = -2
        z = z + 1
```

```python
#print(correlacaoDF)
correlacaoDF.to_csv('correlacaoDF.csv', index=True)
```
